```python
from functools import reduce

'''
13. Dada una lista de Artículos con sus precios. Define las siguientes funciones:
Puedes definir más funciones si lo consideras necesario.
'''
articulos = {
    'Camisa': 20,
    'Pantalón': 30,
    'Calcetines': 5,
    'Zapatos': 50,
    'Gorra': 10,
    'Bufanda': 15,
    'Gafas': 25,
    'Reloj': 35,
    'Corbata': 40,
}

compra = ['Camisa', 'Pantalón', 'Pantalón', 'Gorra', 'Gafas', 'Corbata'] # 155 


# A. Una función que calcule el precio total de la compra

obtener_lista_precios = lambda x: list(map(articulos.get, x))
precio_total = lambda x: reduce(lambda x,y: x+y, obtener_lista_precios(x), 0)

print(f'Precio total de la compra: {precio_total(compra)} €')


# B. Una función que calcule el precio total de la compra con un descuento del 10%

descuento_10 = lambda x: precio_total(x) * 0.9

print(f'Precio total de la compra con 10% de descuento: {descuento_10(compra)} €')


# C. Una función que calcule el precio total de la compra con un descuento del 10% si la compra supera los 100€

descuento_10_mayor_100 = lambda x: precio_total(x) * 0.9 if precio_total(x) > 100 else precio_total(x)

print(f'Precio total de la compra con 10% de descuento si supera 100€: {descuento_10_mayor_100(compra)} €')
print(f'Precio total de la compra con 10% de descuento si supera 100€: {descuento_10_mayor_100(["Camisa", "Pantalón"])} €')


# D. Una función que calcule el precio total aplicando el IVA (21%)
precio_iva = lambda x: round(precio_total(x) * 1.21, 2)

print(f'Precio total de la compra + IVA: {precio_iva(compra)} €')


# E.lista los artículos cuyo precio es superior a 20€ # [Pantalón, Pantalón, Gafas, Corbata]
lista_articulos_20 = lambda x: list(filter(lambda x: articulos[x] > 20, x))

print(f'Lista de artículos con precio superior a 20€: {lista_articulos_20(compra)}')
```

```python
'''
14. Dada una lista de tuplas con el nombre de un alumno, apellidos, curso y sus notas. 

 Define una funcion que reciba el curso y saque una lista en la que aparezca el nombre y apellidos y el promedio 
 de sus notas.

 Puedes definir más funciones si lo consideras necesario.
'''
from functools import reduce

alumnos = [('Juan', 'Pérez', '1', [5, 6, 7, 8, 9]),
            ('Ana', 'García', '2', [5, 6, 7, 8, 9]),
            ('Luis', 'González', '1', [5, 6, 7, 8, 9]),
            ('María', 'Martínez', '2', [5, 6, 7, 8, 9]),
            ('Pedro', 'Rodríguez', '1', [5, 6, 7, 8, 9]),
            ('Lucía', 'Hernández', '2', [5, 6, 7, 8, 9]),
            ('Marta', 'López', '1', [5, 6, 7, 8, 9]),
            ('Sara', 'Díaz', '2', [5, 6, 7, 8, 9]),
            ('Javier', 'Sánchez', '1', [5, 6, 7, 8, 9]),
            ('Miguel', 'Fernández', '2', [5, 6, 7, 8, 9]),
            ('Sergio', 'Jiménez', '1', [5, 6, 7, 8, 9]),
            ('Sandra', 'Ruiz', '2', [5, 6, 7, 8, 9]),
            ('Pablo', 'Álvarez', '1', [5, 6, 7, 8, 9]),
            ('María', 'Gómez', '2', [5, 6, 7, 8, 9]),
]


promedio_notas = lambda notas: reduce(lambda x, y: x + y, notas) / len(notas)

curso = lambda curso: list(filter(lambda x: x[2] == curso, alumnos))

selecciona_curso_alumnos = lambda x: list(map(lambda x: (x[0]+ ' ' + x[1]+ ' ' + str(promedio_notas(x[3]))), curso(x)))

print(f'Lista de alumnos del curso 1: \n{selecciona_curso_alumnos("1")}\n')
print(f'Lista de alumnos del curso 2: \n{selecciona_curso_alumnos("2")}\n')
```
